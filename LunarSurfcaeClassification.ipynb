```python
# Artificial Lunar Rocky Landscape Dataset Analysis
# Enhanced model with improved accuracy techniques

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0, ResNet50V2, Xception
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, LearningRateScheduler
from tensorflow.keras.optimizers import Adam, AdamW
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.regularizers import l2
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
from PIL import Image, ImageDraw, ImageEnhance, ImageOps
import cv2

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Check for GPU availability
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# Set paths
DATASET_PATH = '/kaggle/input/artificial-lunar-rocky-landscape-dataset'
IMAGES_PATH = os.path.join(DATASET_PATH, 'images')
BOUNDING_BOXES_PATH = os.path.join(DATASET_PATH, 'bounding_boxes.csv')
REAL_IMAGES_PATH = os.path.join(DATASET_PATH, 'real_moon_images')

# Configuration parameters
IMG_SIZE = 224  # Standard size for many pre-trained models
BATCH_SIZE = 16  # Reduced batch size to allow more stochasticity
EPOCHS = 50  # Increased epochs for better learning
LEARNING_RATE = 3e-4  # Adjusted learning rate
WEIGHT_DECAY = 1e-5  # L2 regularization factor

# Step 1: Enhanced Dataset Analysis
def analyze_class_distribution():
    """Analyze class distribution in the dataset."""
    print("Analyzing class distribution...")

    class_counts = {}

    if os.path.exists(IMAGES_PATH):
        subdirs = os.listdir(IMAGES_PATH)
        for subdir in subdirs:
            subdir_path = os.path.join(IMAGES_PATH, subdir)
            if os.path.isdir(subdir_path):
                files = os.listdir(subdir_path)
                image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
                class_counts[subdir] = len(image_files)
                print(f"  - Class '{subdir}' has {len(image_files)} images")

    # Plot class distribution
    plt.figure(figsize=(10, 6))
    classes = list(class_counts.keys())
    counts = list(class_counts.values())
    bars = plt.bar(classes, counts, color='skyblue')

    # Add count labels on bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height}', ha='center', va='bottom')

    plt.title('Class Distribution in Lunar Landscape Dataset')
    plt.xlabel('Class')
    plt.ylabel('Number of Images')
    plt.tight_layout()
    plt.show()

    # Calculate class weights for handling imbalance
    total_samples = sum(counts)
    class_weights = {i: total_samples / (len(classes) * count) for i, count in enumerate(counts)}
    print(f"Class weights to handle imbalance: {class_weights}")

    return class_weights

# Step 2: Enhanced Image Preprocessing
def enhance_image(img_array):
    """Apply enhancements to images to improve feature visibility."""
    # Convert to PIL Image
    img = Image.fromarray((img_array * 255).astype(np.uint8))

    # Convert to grayscale and back to RGB for consistency
    # This emphasizes texture over color for lunar surfaces
    img_gray = ImageOps.grayscale(img)
    img_gray_rgb = ImageOps.colorize(img_gray, black="black", white="white")

    # Enhance contrast to make features more visible
    enhancer = ImageEnhance.Contrast(img_gray_rgb)
    img_enhanced = enhancer.enhance(1.5)  # Increase contrast by 50%

    # Apply histogram equalization
    img_np = np.array(img_enhanced)
    if len(img_np.shape) == 3 and img_np.shape[2] == 3:
        # For RGB images, apply equalization to each channel
        img_eq = np.zeros_like(img_np)
        for i in range(3):
            img_eq[:,:,i] = cv2.equalizeHist(img_np[:,:,i])
    else:
        # For grayscale
        img_eq = cv2.equalizeHist(img_np)

    # Normalize back to [0,1] range
    return img_eq / 255.0

# Step 3: Improved Data Preparation with Data Augmentation
def prepare_classification_data_enhanced(img_size=224, batch_size=16, use_enhancement=True):
    """Prepare data with enhanced preprocessing and augmentation."""
    print("\nPreparing enhanced data for classification...")

    # Custom preprocessing function
    def preprocess_input(x):
        if use_enhancement:
            # Apply our custom enhancements
            return enhance_image(x)
        else:
            # Standard preprocessing (just rescaling)
            return x / 255.0

    # Create data generators with more robust augmentation
    train_datagen = ImageDataGenerator(
        preprocessing_function=preprocess_input,
        rotation_range=15,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.15,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='reflect',  # Changed from 'nearest' for smoother interpolation
        validation_split=0.2
    )

    # Simple preprocessing for validation and test
    val_datagen = ImageDataGenerator(
        preprocessing_function=preprocess_input,
        validation_split=0.2
    )

    # Load training data
    train_generator = train_datagen.flow_from_directory(
        IMAGES_PATH,
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    # Load validation data
    validation_generator = val_datagen.flow_from_directory(
        IMAGES_PATH,
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        shuffle=False  # Don't shuffle validation to maintain consistency
    )

    # For testing, we'll use the validation set
    test_generator = validation_generator

    print(f"Found {len(train_generator.class_indices)} classes")
    print(f"Class mapping: {train_generator.class_indices}")
    print(f"Training samples: {train_generator.samples}")
    print(f"Validation samples: {validation_generator.samples}")

    # Visualize some preprocessed images
    x_batch, y_batch = next(train_generator)
    plt.figure(figsize=(15, 6))
    for i in range(min(8, len(x_batch))):
        plt.subplot(2, 4, i+1)
        plt.imshow(x_batch[i])
        class_idx = np.argmax(y_batch[i])
        class_name = list(train_generator.class_indices.keys())[class_idx]
        plt.title(f"Class: {class_name}")
        plt.axis('off')
    plt.suptitle("Enhanced Preprocessed Images")
    plt.tight_layout()
    plt.show()

    return train_generator, validation_generator, test_generator

# Step 4: Alternative Model Architectures
def build_enhanced_classification_model(num_classes, architecture="efficientnet", dropout_rate=0.5, weight_decay=1e-5):
    """Build an enhanced classification model with choice of architectures."""
    print(f"\nBuilding enhanced classification model using {architecture}...")

    if architecture == "efficientnet":
        # EfficientNet with enhanced regularization
        base_model = EfficientNetB0(
            weights='imagenet',
            include_top=False,
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )
    elif architecture == "resnet":
        # Try ResNet50V2 as alternative architecture
        base_model = ResNet50V2(
            weights='imagenet',
            include_top=False,
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )
    elif architecture == "xception":
        # Xception might be good for texture recognition
        base_model = Xception(
            weights='imagenet',
            include_top=False,
            input_shape=(IMG_SIZE, IMG_SIZE, 3)
        )
    else:
        raise ValueError(f"Unknown architecture: {architecture}")

    # Freeze base model layers initially
    for layer in base_model.layers:
        layer.trainable = False

    # Create model with enhanced classification head
    model = Sequential([
        base_model,
        GlobalAveragePooling2D(),
        BatchNormalization(),  # Added BatchNorm for better stability
        Dropout(dropout_rate),  # Increased dropout for stronger regularization
        Dense(256, activation='relu', kernel_regularizer=l2(weight_decay)),  # Added more capacity + L2
        Dropout(dropout_rate),  # Additional dropout layer
        Dense(num_classes, activation='softmax', kernel_regularizer=l2(weight_decay))  # L2 on final layer
    ])

    # Compile model with AdamW optimizer
    model.compile(
        optimizer=AdamW(learning_rate=LEARNING_RATE, weight_decay=weight_decay),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    # Print summary
    model.summary()

    return model, base_model

# Step 5: Learning Rate Scheduler with Warmup
def lr_schedule_with_warmup(epoch, lr):
    """Implements warm-up + cosine decay learning rate schedule."""
    warmup_epochs = 5
    if epoch < warmup_epochs:
        # Linear warm-up
        return LEARNING_RATE * ((epoch + 1) / warmup_epochs)
    else:
        # Cosine decay
        decay_epochs = EPOCHS - warmup_epochs
        decay_progress = (epoch - warmup_epochs) / decay_epochs
        cosine_decay = 0.5 * (1 + np.cos(np.pi * decay_progress))
        return LEARNING_RATE * cosine_decay

# Step 6: Improved Training with Enhanced Callbacks
def train_model_enhanced(model, train_generator, validation_generator, class_weights=None):
    """Enhanced training with better callbacks and strategies."""
    print("\nTraining model with enhanced strategies...")

    # Set up callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=10,  # Increased patience for more training time
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,  # More aggressive reduction
            patience=5,
            min_lr=1e-7,
            verbose=1
        ),
        LearningRateScheduler(lr_schedule_with_warmup, verbose=1),
        ModelCheckpoint(
            'best_lunar_model_enhanced.keras',
            monitor='val_accuracy',
            mode='max',
            save_best_only=True,
            verbose=1
        )
    ]

    # Train the model
    history = model.fit(
        train_generator,
        steps_per_epoch=train_generator.samples // train_generator.batch_size + 1,
        validation_data=validation_generator,
        validation_steps=validation_generator.samples // validation_generator.batch_size + 1,
        epochs=EPOCHS,
        callbacks=callbacks,
        class_weight=class_weights,
        verbose=1
    )

    return history

# Add the missing function: Plot training history
def plot_training_history(history, title="Training History"):
    """Plot training & validation accuracy and loss."""
    # Plot training & validation accuracy values
    plt.figure(figsize=(12, 5))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title(f'{title} - Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='lower right')

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title(f'{title} - Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper right')

    plt.tight_layout()
    plt.show()

    # Print highest validation accuracy
    val_acc = history.history['val_accuracy']
    max_val_acc = max(val_acc)
    max_idx = val_acc.index(max_val_acc)
    print(f"Highest validation accuracy: {max_val_acc:.4f} (epoch {max_idx+1})")

# Step 7: Enhanced Evaluation with Feature Visualization
def evaluate_model_enhanced(model, test_generator, base_model=None):
    """Comprehensive model evaluation with feature visualization."""
    print("\nComprehensive model evaluation...")

    # Get predictions
    test_generator.reset()
    predictions = model.predict(test_generator)
    predicted_classes = np.argmax(predictions, axis=1)

    # Get true classes
    true_classes = test_generator.classes[:len(predicted_classes)]

    # Get class labels
    class_labels = list(test_generator.class_indices.keys())

    # Print classification report
    print("\nClassification Report:")
    report = classification_report(true_classes, predicted_classes, target_names=class_labels)
    print(report)

    # Plot confusion matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_labels, yticklabels=class_labels)
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.show()

    # Analyze misclassifications
    print("\nAnalyzing misclassifications...")
    misclassified_indices = np.where(predicted_classes != true_classes)[0]
    if len(misclassified_indices) > 0:
        # Display some misclassified examples
        plt.figure(figsize=(15, 10))
        test_generator.reset()
        x_test, y_test = next(test_generator)
        for i, idx in enumerate(misclassified_indices[:min(8, len(misclassified_indices))]):
            if idx < len(x_test):
                plt.subplot(2, 4, i+1)
                plt.imshow(x_test[idx])
                true_label = class_labels[np.argmax(y_test[idx])]
                pred_label = class_labels[np.argmax(predictions[idx])]
                plt.title(f"True: {true_label}\nPred: {pred_label}")
                plt.axis('off')
        plt.suptitle("Misclassified Examples")
        plt.tight_layout()
        plt.show()

    # Feature visualization
    if base_model is not None:
        # Create a feature extraction model
        feature_model = Model(inputs=base_model.input, outputs=base_model.output)

        # Get a batch of images
        test_generator.reset()
        x_batch, _ = next(test_generator)

        # Extract features
        features = feature_model.predict(x_batch[:4])  # Just look at a few examples

        # Visualize feature maps (first layer only for demonstration)
        plt.figure(figsize=(15, 10))
        for i in range(min(4, len(x_batch))):
            # Display original image
            plt.subplot(2, 4, i+1)
            plt.imshow(x_batch[i])
            plt.title(f"Original {i+1}")
            plt.axis('off')

            # Display feature map (average across channels)
            plt.subplot(2, 4, i+5)
            feature_map = np.mean(features[i], axis=-1)
            plt.imshow(feature_map, cmap='viridis')
            plt.title(f"Feature Map {i+1}")
            plt.axis('off')

        plt.suptitle("Feature Visualization")
        plt.tight_layout()
        plt.show()

    # Calculate accuracy
    accuracy = (predicted_classes == true_classes).mean()
    print(f"Test Accuracy: {accuracy * 100:.2f}%")

    return accuracy, report

# Step 8: Cross-Validation for Better Generalization
def cross_validation_evaluation(architecture="efficientnet", n_folds=5):
    """Evaluate model using stratified k-fold cross-validation."""
    print(f"\nPerforming {n_folds}-fold cross-validation...")

    # Load all data
    datagen = ImageDataGenerator(preprocessing_function=lambda x: x/255.0)
    all_generator = datagen.flow_from_directory(
        IMAGES_PATH,
        target_size=(IMG_SIZE, IMG_SIZE),
        batch_size=BATCH_SIZE,
        class_mode='categorical',
        shuffle=False
    )

    # Get all images and labels
    all_generator.reset()
    x_all = []
    y_all = []

    # This could take a while for large datasets
    steps = all_generator.samples // all_generator.batch_size + 1
    for i in range(steps):
        batch_x, batch_y = next(all_generator)
        x_all.append(batch_x)
        y_all.append(batch_y)

    x_all = np.vstack(x_all)
    y_all = np.vstack(y_all)

    # Get class indices
    class_indices = all_generator.class_indices

    # Create stratified k-fold
    skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

    # Convert categorical y to class indices for stratification
    y_indices = np.argmax(y_all, axis=1)

    # Track accuracies for each fold
    fold_accuracies = []

    # For each fold
    for fold, (train_idx, val_idx) in enumerate(skf.split(x_all, y_indices)):
        print(f"\nFold {fold+1}/{n_folds}")

        # Split data
        x_train, x_val = x_all[train_idx], x_all[val_idx]
        y_train, y_val = y_all[train_idx], y_all[val_idx]

        # Build model
        num_classes = len(class_indices)
        model, base_model = build_enhanced_classification_model(
            num_classes=num_classes,
            architecture=architecture,
            dropout_rate=0.5,
            weight_decay=WEIGHT_DECAY
        )

        # Create custom data generators
        train_datagen = ImageDataGenerator(
            rotation_range=15,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.15,
            zoom_range=0.2,
            horizontal_flip=True,
            vertical_flip=True,
            fill_mode='reflect'
        )

        # Train generator with augmentation
        train_generator = train_datagen.flow(
            x_train, y_train,
            batch_size=BATCH_SIZE,
            shuffle=True
        )

        # Validation generator without augmentation
        val_generator = ImageDataGenerator().flow(
            x_val, y_val,
            batch_size=BATCH_SIZE,
            shuffle=False
        )

        # Train for fewer epochs in cross-validation
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-7)
        ]

        # Train model
        history = model.fit(
            train_generator,
            steps_per_epoch=len(x_train) // BATCH_SIZE + 1,
            validation_data=val_generator,
            validation_steps=len(x_val) // BATCH_SIZE + 1,
            epochs=15,  # Fewer epochs for CV
            callbacks=callbacks,
            verbose=1
        )

        # Evaluate
        scores = model.evaluate(val_generator)
        print(f"Fold {fold+1} - Validation Accuracy: {scores[1]*100:.2f}%")
        fold_accuracies.append(scores[1])

        # Clear memory
        tf.keras.backend.clear_session()

    # Print overall results
    print("\nCross-Validation Results:")
    print(f"Mean Accuracy: {np.mean(fold_accuracies)*100:.2f}%")
    print(f"Standard Deviation: {np.std(fold_accuracies)*100:.2f}%")

    return np.mean(fold_accuracies), np.std(fold_accuracies)

# Step 9: Enhanced Classification Pipeline
def run_enhanced_classification_pipeline():
    """Run the enhanced classification pipeline."""
    print("\nRunning enhanced classification pipeline...")

    # Analyze class distribution first
    class_weights = analyze_class_distribution()

    # Prepare enhanced data
    train_generator, validation_generator, test_generator = prepare_classification_data_enhanced(
        img_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        use_enhancement=True  # Apply image enhancements
    )

    # Get number of classes
    num_classes = len(train_generator.class_indices)

    # Try different architectures
    architectures = ["efficientnet", "resnet", "xception"]
    best_accuracy = 0
    best_model = None
    best_architecture = None

    for arch in architectures:
        print(f"\n===== Evaluating {arch} architecture =====")

        # Build model
        model, base_model = build_enhanced_classification_model(
            num_classes=num_classes,
            architecture=arch,
            dropout_rate=0.5,  # High dropout for better regularization
            weight_decay=WEIGHT_DECAY
        )

        # Train model
        history = train_model_enhanced(
            model,
            train_generator,
            validation_generator,
            class_weights=class_weights  # Use class weights to handle imbalance
        )

        # Plot training history
        plot_training_history(history, f"Training History - {arch}")

        # Evaluate model
        accuracy, _ = evaluate_model_enhanced(model, test_generator, base_model)

        # Track best model
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_model = model
            best_architecture = arch

            # Save best model
            model.save(f'best_lunar_model_{arch}.keras')

    print(f"\nBest architecture: {best_architecture} with accuracy: {best_accuracy*100:.2f}%")

    # Fine-tune the best model
    if best_model is not None:
        print(f"\nFine-tuning the best model ({best_architecture})...")

        # Gradually unfreeze layers and fine-tune
        model = best_model
        base_model = model.layers[0]

        # Unfreeze the entire base model
        for layer in base_model.layers:
            layer.trainable = True

        # Compile with lower learning rate
        model.compile(
            optimizer=AdamW(learning_rate=LEARNING_RATE/10, weight_decay=WEIGHT_DECAY),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )

        # Fine-tune with reduced learning rate
        fine_tune_history = model.fit(
            train_generator,
            steps_per_epoch=train_generator.samples // BATCH_SIZE + 1,
            validation_data=validation_generator,
            validation_steps=validation_generator.samples // BATCH_SIZE + 1,
            epochs=20,
            callbacks=[
                EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
                ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-7),
                ModelCheckpoint(f'best_lunar_model_{best_architecture}_finetuned.keras',
                               monitor='val_accuracy', mode='max', save_best_only=True)
            ],
            class_weight=class_weights,
            verbose=1
        )

        # Plot fine-tuning history
        plot_training_history(fine_tune_history, f"Fine-Tuning History - {best_architecture}")

        # Final evaluation
        final_accuracy, _ = evaluate_model_enhanced(model, test_generator)
        print(f"\nFinal model accuracy after fine-tuning: {final_accuracy*100:.2f}%")

    # Cross-validation evaluation on best architecture
    print(f"\nPerforming cross-validation with {best_architecture}...")
    cv_accuracy, cv_std = cross_validation_evaluation(best_architecture)

    print("\nClassification pipeline completed!")
    return best_architecture, best_accuracy

# Main function
def main():
    print("=" * 50)
    print("ENHANCED LUNAR LANDSCAPE ANALYSIS")
    print("=" * 50)

    # Run enhanced classification pipeline
    best_arch, best_acc = run_enhanced_classification_pipeline()

    print(f"\nBest solution: {best_arch} architecture with {best_acc*100:.2f}% accuracy")
    print("=" * 50)
    print("ANALYSIS COMPLETE")
    print("=" * 50)

# Additional strategies to improve model performance (can be added to the pipeline)
def strategies_to_improve_model():
    """Additional strategies that can be implemented to improve model performance."""

    print("\nAdditional strategies to improve model performance:")
    print("1. Implement Mixup or CutMix data augmentation for better generalization")
    print("2. Try attention mechanisms (e.g., SE blocks) to focus on important features")
    print("3. Implement Stochastic Weight Averaging (SWA) for better convergence")
    print("4. Experiment with different optimizers like LAMB or SAM")
    print("5. Implement Label Smoothing to prevent overconfidence")
    print("6. Try Knowledge Distillation from larger models")
    print("7. Implement ensemble methods combining different architectures")
    print("8. Use progressive resizing (train on smaller images first, then fine-tune on larger ones)")
    print("9. Try cyclic learning rates or one-cycle policy")
    print("10. Implement test-time augmentation for better inference")

# Run the main function
if __name__ == "__main__":
    main()
```
